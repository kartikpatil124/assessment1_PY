```python

from datetime import date



print("welcome to the interactive personal data collector")

name = input("Enter your name")
age = int(input("Enter your age"))
height = float(input("Enter your height"))
num = input("Enter your Favourite Number")



print(f"Your Name is {name}")
print(f"Your age is {age}")
print(f"Your height is {height}")
print(f"Your Favourite Number is {num}")

print("Thank You,here is the information we collected:")

print("the id for name",(name),":" ,id(name))
print("the id for age",(age),":" ,id(age))
print("the id for height",(height),":" ,id(height))
print("the id for Favourite Number",(num),":" ,id(num))


# take input
year = int(input("Enter birth year (YYYY): "))
month = int(input("Enter birth month (MM): "))
day = int(input("Enter birth day (DD): "))

birth_date = date(year, month, day)
today = date.today()


# calculate age
age = today.year - birth_date.year


# adjust if birthday not yet passed this year
if (today.month, today.day) < (birth_date.month, birth_date.day):
    age -= 1



print("Your age is:", age, "years")
```
